# assignment 3: work with iris

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset('iris')
```

```python
iris
```

```python
sns.scatterplot(x = "sepal_length",y = "sepal_width",data = iris)
```

```python
plt.scatter(iris["sepal_length"],iris["sepal_width"])
```

```python
sns.lineplot(x= "species",y="sepal_width",data = iris)
```

```python
sns.distplot(iris["sepal_width"])
```

```python
sns.barplot(x = "petal_width",y = "sepal_width",data = iris)
```

```python
sns.boxplot(x = "petal_width",y = "sepal_width",data = iris)
```

```python
sns.catplot(x="species",y = "sepal_width",data = iris)
```

```python
sns.pairplot(iris)
```

```python
sns.heatmap(iris.corr(),annot = True)
```
